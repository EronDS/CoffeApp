```python
import os
def classes_to_binary(from_folder:str, to_folder:str, map:dict) -> None:
    os.makedirs(to_folder, exist_ok = True)
    for file_dir in os.listdir(from_folder):
        with open(from_folder+'/'+file_dir, 'r') as f:
            new_file_dir = to_folder + '/' + file_dir
            with open(new_file_dir, 'w') as w:
                for line in f:
                    old_line = line 
                    new_line = map[line[0]] + old_line[1:]
                    w.write(new_line)
```

```python
classes_to_binary(from_folder ='/content/drive/MyDrive/yolov7/data/coffeAI-firstgen-5to2-unsup/train/labels', 
                  to_folder = '/content/drive/MyDrive/yolov7/data/coffeAI-firstgen-5to2-unsup/train/labels_',
                  map = {'0':'1',
                         '1':'0',
                         '2':'0',
                         '3':'1',
                         '4':'0'})
```
